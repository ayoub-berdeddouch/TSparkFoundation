Calculating the IQR: **I**nter**Q**uartile **R**ange

```python

Q1 = df_retail.quantile(q = 0.25, axis = 0, numeric_only = True, interpolation = 'linear')

Q3 = df_retail.quantile(q = 0.75, axis = 0, numeric_only = True, interpolation = 'linear')

IQR = Q3 - Q1

print(IQR)
```

```python
df_retail.value_counts().nlargest().plot(kind = 'bar' , figsize = (10 , 5))
```

```python
print(df_retail['Sales'].describe())
plt.figure(figsize = (9 , 8))
sns.distplot(df_retail['Sales'], color = 'r', bins = 100, hist_kws = {'alpha': 0.4});
print("---------------------------------------------")
print("Skewness: %f" % df_retail['Sales'].skew())
print("Kurtosis: %f" % df_retail['Sales'].kurt())
```

```python
# Log of Sales

plt.figure(figsize = (9 , 8))
sns.distplot(np.log(df_retail['Sales']), color = 'b', bins = 100, hist_kws = {'alpha': 0.4});
print("Skewness: %f" % np.log(df_retail['Sales']).skew())
print("Kurtosis: %f" % np.log(df_retail['Sales']).kurt())
```

```python
print(df_retail['Profit'].describe())
plt.figure(figsize = (9 , 8))
sns.distplot(df_retail['Profit'], color = 'r', bins = 100, hist_kws = {'alpha': 0.4});
print("---------------------------------------------")
print("Skewness: %f" % df_retail['Profit'].skew())
print("Kurtosis: %f" % df_retail['Profit'].kurt())
```

```python
plt.figure(figsize = (9 , 8))
sns.distplot(np.log1p(df_retail['Profit']), color = 'b', bins = 100, hist_kws = {'alpha': 0.4});
print("---------------------------------------------")
print("Skewness: %f" % np.log1p(df_retail['Profit']).skew())
print("Kurtosis: %f" % np.log1p(df_retail['Profit']).kurt())
```

```python
print(df_retail['Discount'].describe())
plt.figure(figsize = (9 , 8))
sns.distplot(df_retail['Discount'], color = 'r', bins = 100, hist_kws = {'alpha': 0.4});
print("---------------------------------------------")
print("Skewness: %f" % df_retail['Discount'].skew())
print("Kurtosis: %f" % df_retail['Discount'].kurt())
```

```python
plt.figure(figsize = (9 , 8))
sns.distplot(np.log1p(df_retail['Discount']), color = 'b', bins = 100, hist_kws = {'alpha': 0.4});
print("---------------------------------------------")
print("Skewness: %f" % np.log1p(df_retail['Discount']).skew())
print("Kurtosis: %f" % np.log1p(df_retail['Discount']).kurt())
```
